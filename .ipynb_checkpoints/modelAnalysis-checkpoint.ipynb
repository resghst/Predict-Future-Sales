```python
from scipy.stats import norm
from scipy import stats

shop = df_train[df_train['shop_id'] == 10]
for i in range(len(shop.columns)): 
    var = shop.columns[i]
    print(var)
    if(var != 'item_cnt_day' and var != 'date' and var != 'shop_id'):
        data = pd.concat([shop['item_cnt_day'], shop[var]], axis=1)
        slope, intercept, r_value, p_value, std_err = stats.linregress(shop['item_cnt_day'], shop[var])
#         print(stats.linregress(shop[var], shop['item_cnt_day']))
        print(data.plot(shop['item_cnt_day'], intercept + slope*shop['item_cnt_day']))
        print(data.plot.scatter(x='item_cnt_day', y=var))
```
